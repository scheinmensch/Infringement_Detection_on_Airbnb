```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
bcn_orginal = pd.read_csv('../data/listings_bcn.csv')
```

```python
bcn_orginal.head()
```

```python
urls = bcn_orginal['listing_url']
```

```python
data=[]
driver = webdriver.Chrome()
for url in urls:
    driver.get(url)
    time.sleep(5)
    html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    isPrivate = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML.match('Private:r Gastgeber:in')")
    notFound = driver.execute_script("return document.title === '410 Page Not Found - Airbnb'")
    if notFound:
        data.append({'url':url,'host_type':'not_found'})
    elif isPrivate:
        data.append({'url':url,'host_type':'private'})
    else:
        data.append({'url':url,'host_type':'commercial'})
driver.quit()
```

```python
df = pd.DataFrame(data) 
```

```python
df.to_csv('airbnb.csv') 
```
